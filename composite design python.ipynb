## **Designing Large Particle Composites**
During the designing of large-particle composites, the dimensions of the particles should be same in all directions i.e it should be of equi-axed nature inspite of possessing different geometries. Below code is used to obtain the elastic modulus for the given composite material if the elastic modulus of the both phases i.e. the matrix phase and the particulate phase is known. In addition to the elastic modulus of both phases volume fraction of these two phases should be also known. 

```
#For two-phase composite, modulus of elasticity of upper bound condition
inp1=input('Enter the value of the Elastic Modulus of Matrix used:')
inp2=input('Enter the value of the Volume Fraction of the Matrix used:')
inp3=input('Enter the value of the Elastic Modulus of the Particulate used:')
inp4=input('Enter the value of the Volume Fraction of the Particulate used:')
try:
  elastic_modulus_matrix=float(inp1)
  volume_fraction_matrix=float(inp2)
  elastic_modulus_particulate=float(inp3)
  volume_fraction_particulate=float(inp4)
  modulus_of_elasticity_upper_bound=((elastic_modulus_matrix*volume_fraction_matrix)+(elastic_modulus_particulate*volume_fraction_particulate))
  print(modulus_of_elasticity_upper_bound)
except:
   print('Please enter a number') 

```

```
#For two-phase, modulus of elasticity of lower bound condition
inp1=input('Enter the value of the Elastic Modulus of Matrix used:')
inp2=input('Enter the value of the Volume Fraction of the Matrix used:')
inp3=input('Enter the value of the Elastic Modulus of the Particulate used:')
inp4=input('Enter the value of the Volume Fraction of the Particulate used:')
try:
  elastic_modulus_matrix=float(inp1)
  volume_fraction_matrix=float(inp2)
  elastic_modulus_particulate=float(inp3)
  volume_fraction_particulate=float(inp4)
  modulus_of_elasticity_lower_bound=((elastic_modulus_matrix*elastic_modulus_particulate)/((volume_fraction_matrix*elastic_modulus_particulate)+(volume_fraction_particulate*elastic_modulus_matrix)))
  print(modulus_of_elasticity_lower_bound)
except:
  print('Please enter a number')  
```

# **Designing Fibre Reinforced Composites**
In this type of composite, dispersed phase is in the form of the fibers. This type of composite is sub-classified according to fiber length. The next code line is to obtain the critical fiber length which is dependent on the diameter of fibre, its ultimate or tensile strength and also depends on the shear yield strength of the matrix. It should be noted that the critical fiber length plays an important role in stiffening and effective strengthning of the composite material.  

```
inp5=input('Enter the value of Ultimate Tensile Strength of the fibre in N/mm^2:')
inp6=input('Enter the value of diameter of fiber in mm:')
inp7=input('Enter the value of Shear yield strength of the matrix in N/mm^2:')
try:
  ultimate_tensile_strength=float(inp5)
  fiber_diameter=float(inp6)
  shear_yield_strength=float(inp7)
  critical_fiber_length=((ultimate_tensile_strength*fiber_diameter)/(2*shear_yield_strength))
  print(critical_fiber_length)
except:
  print('Please enter a number')  
  



```

```
#Checking whether fiber chosen is continuous or discontinuous/short fiber
inp8=input('Enter the length of the available fiber in mm:')
try:
  fiber_length=float(inp8)
  if fiber_length>15*critical_fiber_length:
      print('The fiber chosen is continuous fiber')
  else:
    print('The fiber chosen is discontinuous/short')
except:
  print('Please enter a number')
```
